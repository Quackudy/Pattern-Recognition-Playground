```python
import numpy as np
import pandas as pd
import plotly.express as px

size = 70
steps = 100

grid = np.random.randint(0, 2, (size, size))

def step(grid):
    neighbors = sum(np.roll(np.roll(grid, i, 0), j, 1)
                    for i in (-1,0,1)
                    for j in (-1,0,1)
                    if not (i==0 and j==0))

    return ((neighbors == 3) | ((grid == 1) & (neighbors == 2))).astype(int)

frames = np.zeros((steps, size, size), dtype=int)

for t in range(steps):
    frames[t] = grid
    grid = step(grid)


fig = px.imshow(frames, animation_frame=0)

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 100
fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 0

fig.show()
```
